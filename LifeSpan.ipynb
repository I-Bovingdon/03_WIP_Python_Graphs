```python
# ===================== 
# Importación de librerías y carga de datos
# =====================

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter=';',
    decimal=','
)
```

```python
# =====================
# Ordenación por población
# =====================

df_order = df.sort_values('Poblacion', ascending=False)
df_order
```

```python
# =====================
# Gráfico básico: Renta per cápita vs Esperanza de vida
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)

plt.show()
```

```python
# =====================
# Gráfico con título y etiquetas
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

plt.show()
```

```python
# =====================
# Normalización de la población
# =====================

df_order['Poblacion_normalizada'] = (
    df_order['Poblacion'] / (max(df_order['Poblacion']) / 10000)
)

df_order
```

```python
# =====================
# Gráfico con tamaño proporcional a la población
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

plt.show()
```

```python
# =====================
# Ajuste del tamaño del gráfico
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

plt.show()
```

```python
# =====================
# Tamaño y color según población
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

plt.show()
```

```python
# =====================
# Anotación de los 10 países con mayor población
# =====================

df_order.reset_index(drop=True, inplace=True)

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )

plt.show()
```

```python
# =====================
# Gráfico final con transparencia y barra de color
# =====================

plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'],
    alpha=0.5
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

plt.colorbar()

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

df_order.reset_index(drop=True, inplace=True)

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )

plt.show()
```
